```python
#food consumption distribution
plt.figure(figsize=(10,8))
sns.set_style("darkgrid")
sns.kdeplot(data=my_data["consumption"], shade=True)
```

```python
group_co2_emmission = my_data.groupby('food_category')['co2_emmission'].sum().sort_values()
plt.subplots(figsize=(15,8))
group_co2_emmission.plot(kind='barh', fontsize=12)
plt.xlabel('amount of emmision', fontsize=12)
plt.ylabel('food category', fontsize=12)
plt.title('amount of co2 emmision for different categories', fontsize=12)
plt.show()
```

```python
cons = my_data.dropna(subset = ['consumption'])
cons['country'].value_counts().nlargest(5).plot(kind = 'bar', linewidth = 1, facecolor = 'seagreen', edgecolor = 'k')
#here sum method is not working, so I had to use value_counts
plt.title('top 5 countries with most consumption level')
plt.xlabel('country')
plt.ylabel('consumption level')
plt.show()
```
